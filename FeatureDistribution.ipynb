```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate some data
np.random.seed(42)

# Class 0: non-defaulters
x1_0 = np.random.normal(35, 10, 500)  # age
x2_0 = np.random.normal(50000, 15000, 500)  # income

# Class 1: defaulters
x1_1 = np.random.normal(28, 8, 500)
x2_1 = np.random.normal(30000, 10000, 500)

df = pd.DataFrame({
    "X1": np.concatenate([x1_0, x1_1]),
    "X2": np.concatenate([x2_0, x2_1]),
    "Y": [0]*500 + [1]*500
})
```

```python
sns.kdeplot(data=df, x="X1", y="X2", hue="Y", fill=True, alpha=0.5)
plt.title("Estimated P(X1, X2) for Each Class")
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```

Orange: Defaulters, blau: Non-defaulters.  Je mehr die Wolen sich überlappen, desto weniger aussagekräftig sind die Features. 

```python
sns.scatterplot(data=df, x="X1", y="X2", hue="Y", alpha=0.6)
plt.title("Scatter of Age vs Income by Default Status")
```
