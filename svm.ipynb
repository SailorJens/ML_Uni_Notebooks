```python
import numpy as np
import matplotlib.pyplot as plt

# Create a range of decision values
decision_values = np.linspace(-2, 2, 400)
labels = np.array([1])  # Assume the label y = 1

# Hinge loss for y = 1
hinge_loss = np.maximum(0, 1 - labels * decision_values)
squared_hinge_loss = hinge_loss ** 2

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(decision_values, hinge_loss, label='Hinge Loss', color='blue')
plt.plot(decision_values, squared_hinge_loss, label='Squared Hinge Loss', color='red', linestyle='--')
plt.axvline(x=1, color='gray', linestyle=':', label='Margin Boundary (y=1)')
plt.axvline(x=0, color='black', linestyle='--', label='Decision Boundary')
plt.title("Hinge Loss vs. Squared Hinge Loss (for y=1)")
plt.xlabel("Decision Function Output (wx + b)")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
